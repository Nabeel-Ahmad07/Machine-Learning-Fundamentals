### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, classification_report
```

### Load the Cancer Dataset

```python
cancer = load_breast_cancer()
X = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y = pd.Series(cancer.target)

print("Dataset shape:", X.shape)
print("Target classes:", cancer.target_names)
# print(X.head)
# print(y.head)
```

### Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)
print("Done Splitting")
```

### Train Decision Tree (maxDepth = 3)

```python
clf = DecisionTreeClassifier(max_depth=3, random_state=87)
clf.fit(X_train, y_train)
```

### Training and Testing Accuracy with Confusion Matrix

```python
y_pred = clf.predict(X_test)
print("Training Accuracy:", clf.score(X_train, y_train))
print("Testing Accuracy:", accuracy_score(y_test, y_pred))

ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=cancer.target_names, cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```

### Visualizing Decision Tree

```python
plt.figure(figsize=(15, 8))
plot_tree(clf, feature_names=X.columns, class_names=cancer.target_names, filled=True, fontsize=8)
plt.title("Decision Tree")
plt.show()
```

### Classification Report

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```
