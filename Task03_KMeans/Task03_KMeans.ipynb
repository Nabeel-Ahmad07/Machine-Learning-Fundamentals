### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

### Load Iris Dataset

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)          
y = pd.Series(iris.target)
print("Dataset Shape: ", X.shape)

# print(X.head())
# print(X.info())
```

### Apply KMeans 

```python
kmeans = KMeans(n_clusters=3, random_state=99)
y_kmeans = kmeans.fit_predict(X)
```

### Confusion Matrix

```python
ConfusionMatrixDisplay.from_predictions(y, y_kmeans, cmap="Blues", display_labels=iris.target_names)
plt.title("Confusion Matrix")
plt.show()
```

### Classification Report

```python
print("Classification Report:\n", classification_report(y, y_kmeans, target_names=iris.target_names))
```

### Plot clusters with 2 features

```python
# here we are doing it on the basis of sepal length and width
# print(X.columns)
col1 = X.columns[0]
col2 = X.columns[1] 

plt.scatter(X[col1], X[col2], c=y_kmeans, cmap='plasma', s=20)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],c='black', marker='X', s=200)
plt.xlabel(col1)
plt.ylabel(col2)
plt.title("KMeans")
plt.show()
```
